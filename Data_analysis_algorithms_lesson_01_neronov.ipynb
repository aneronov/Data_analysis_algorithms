1. Подберите скорость обучения (eta) и количество итераций

```python
import numpy as np
```

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
n = X.shape[0]
eta = 1e-2 
#ИЗМЕНЕНИЯ
n_iter = 700 #изменил значение со 100 на 000
#ИЗМЕНЕНИЯ
W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = np.mean((y - y_pred)**2)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
        #ИЗМЕНЕНИЯ
        eta /= 1.00  #измениел делитель с 1.1 на 1.0
        #ИЗМЕНЕНИЯ
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

3*. Вместо того, чтобы задавать количество итераций, задайте другое условие останова алгоритма - когда веса перестают изменяться меньше определенного порога $\epsilon$.

```python
n = X.shape[0]

eta = 1e-2 
n_iter = 700
w_out = [0, 0]

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = np.mean((y - y_pred)**2)
    W -= eta * (1/n * 2 * np.dot(X.T, y_pred - y))
    if i % 10 == 0:
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err,4)}')
    
    if np.linalg.norm(W - w_out, ord=2) < 0.001:
        break
    w_out = W.copy()
```
